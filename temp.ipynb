```python
# 方案1：检查 HuggingFace 缓存目录，看是否真的在下载
import os
from pathlib import Path

cache_dir = Path.home() / ".cache" / "huggingface" / "datasets"
print(f"缓存目录: {cache_dir}")
print(f"缓存目录是否存在: {cache_dir.exists()}")

if cache_dir.exists():
    # 列出最近修改的文件，看是否有新文件在增长
    import time
    files = list(cache_dir.rglob("*"))
    if files:
        recent_files = sorted(files, key=lambda x: x.stat().st_mtime if x.is_file() else 0, reverse=True)[:5]
        print("\n最近修改的5个文件:")
        for f in recent_files:
            if f.is_file():
                size_mb = f.stat().st_size / 1024 / 1024
                mtime = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(f.stat().st_mtime))
                print(f"  {f.name[:50]}: {size_mb:.2f}MB, 修改时间: {mtime}")
    else:
        print("缓存目录为空")
else:
    print("缓存目录不存在，可能还没开始下载")
```
